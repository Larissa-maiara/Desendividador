```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from google.colab import drive

drive.mount('/content/drive', force_remount=True)

# Caminho do arquivo
caminho = "/content/drive/My Drive/taxascredito - base dados para juros.xlsx"

# Carrega a aba unificado
df = pd.read_excel(caminho, sheet_name="unificado")

# Limpeza dos nomes das colunas
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace('\xa0','')
df.columns = df.columns.str.replace('\t','')

col = "TAXAS MÉDIAS % a.m."

# ================================
# LIMPAR E CONVERTER A COLUNA
# ================================
# Remove espaços e converte para número
df[col] = (
    df[col]
    .astype(str)        # transforma em string
    .str.replace(',', '.')   # troca vírgula por ponto, se houver
    .str.strip()             # remove espaços
)

# Converte realmente para float; valores inválidos viram NaN
df[col] = pd.to_numeric(df[col], errors='coerce')

print("Tipos após conversão:")
print(df[col].dtype)

# Filtra linhas desejadas
dados = df.loc[2:818, [col]].dropna()

# Padronização
scaler = StandardScaler()
dados_scaled = scaler.fit_transform(dados)

# KMeans com 4 clusters
kmeans = KMeans(n_clusters=4, random_state=42)
cluster_labels = kmeans.fit_predict(dados_scaled)

dados["Cluster"] = cluster_labels

dados.head()
```

```python
# Ordena os clusters pela média de juros (do menor para o maior)
faixas = (
    dados.groupby("Cluster")["TAXAS MÉDIAS % a.m."]
    .agg(["min", "max", "mean"])
    .sort_values("mean")
)

faixas
```
